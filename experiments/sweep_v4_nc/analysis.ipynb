```python
import pandas as pd
```

```python
results = pd.read_json("result.jsonl", lines=True)

records = []
for _, row in results.iterrows():
    record = {
        "lr": row["args"]["lr"],
        "wd": row["args"]["wd"],
        "acc_train": row["acc_train"],
        "acc_val": row["acc_validation"],
        "acc_test": row["acc_test"],
    }
    records.append(record)
df = pd.DataFrame.from_records(records)

print(df.shape)
df.head()
```
