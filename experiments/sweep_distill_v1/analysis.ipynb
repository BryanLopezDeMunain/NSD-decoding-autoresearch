```python
import pandas as pd
```

```python
results = pd.read_json("results.jsonl", lines=True)

records = []
for _, row in results.iterrows():
    record = {
        "subs": row["args"]["subs"],
        "lr": row["args"]["lr"],
        "wd": row["args"]["wd"],
        "acc_train": row["acc_train"],
        "acc_val": row["acc_val"],
        "acc_test": row["acc_test"],
    }
    records.append(record)
df = pd.DataFrame.from_records(records)

print(df.shape)
df.head()
```

```python
summary = []
for subs, group in df.groupby("subs"):
    idx = group["acc_val"].idxmax()
    summary.append(group.loc[idx])
summary = pd.DataFrame.from_records(summary)
summary
```

```python
summary_fmt = summary.copy()
summary_fmt = summary_fmt.loc[[0, 2, 3, 1], ["subs", "acc_val", "acc_test"]]
summary_fmt.columns = ["subjects", "val (testid)", "test (shared1k)"]
print(summary_fmt.to_markdown(index=False))
```
